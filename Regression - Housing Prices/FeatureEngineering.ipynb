```python
train_data = pd.read_csv('train.csv')
train_data.shape
```

```python
train_data.head()
```

```python
train_data.pop('Id')
train_data_responses = train_data.pop('SalePrice')
```

```python
#Now I can concat the train and test data to apply the transformations at the same time
test_data = pd.read_csv('test.csv')
#Required for submission
test_data_id = test_data.pop('Id')
test_data.shape
```

```python
dataset = pd.concat((train_data,test_data))
dataset.shape
```

### Dropping the Alley, PoolQC, Fence,MiscFeature columns since they are mostly NaN

```python
dataset = dataset.drop(columns=['Alley','PoolQC','Fence','MiscFeature'])
```

### Handling NaN values

```python
#For categorical variables
cat_variables = [f for f in dataset.columns if dataset[f].dtypes=='object']
#Replacing the nan values with a new category to use dummies later
for i in cat_variables:
    dataset[i] = dataset[i].fillna('XXX')
dataset[cat_variables].head()
```

```python
#MSSubClass is also categorical by data description
dataset['MSSubClass'].isna().sum()
```

```python
dataset[cat_variables].isna().sum()
```

### Categorical variable mapping

```python

for i in cat_variables:
    dataset[i] = dataset[i].astype('category')
    dataset[i] = dataset[i].cat.codes
dataset.head(20)
```

### Numerical Variables

```python
num_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'object' ]
dataset[num_features].isna().sum()
```

```python
#Garage year built cant be earlier than the year built, so, making nan values equal to year built.
d1 = dataset.copy()
d1['GarageYrBlt'].fillna(d1['YearBuilt'], inplace=True)
dataset['GarageYrBlt'] = d1['GarageYrBlt']
```

```python
num_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'object' ]
dataset[num_features].isna().sum()
```

```python
#Too many outliers in the data, so will replace nan values with median instead of mean
#Creating a new feautre for nan values of LotFrontage and MasVnrArea to indicate they are missing. Rest 
# is too small so I'm not doing for them

dataset['LotFrontage'+'NaN'] = np.where(dataset['LotFrontage'].isna(),1,0)
dataset['MasVnrArea'+'NaN'] = np.where(dataset['MasVnrArea'].isna(),1,0)

for f in num_features:
    median = dataset[f].median()    
    dataset[f].fillna(median,inplace=True)
```

### Feature Scaling

```python
#Continuous values were skewed so log transformation:
con_feature = ['LotFrontage',
 'LotArea',
 'MasVnrArea',
 'BsmtFinSF1',
 'BsmtFinSF2',
 'BsmtUnfSF',
 'TotalBsmtSF',
 '1stFlrSF',
 '2ndFlrSF',
 'GrLivArea','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','ScreenPorch']
```

```python
final = dataset.copy()
for i in con_feature:
    #preventing log(0)
    final[i] += 0.0000000001
    final[i] = np.log(final[i])
```

```python
final.shape
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

final[final.columns] = scale.fit_transform(final)
final.head()
```

```python
x_train = final[:1460]
x_train.shape
```

```python
x_test = final[1460:]
x_test.shape
```
